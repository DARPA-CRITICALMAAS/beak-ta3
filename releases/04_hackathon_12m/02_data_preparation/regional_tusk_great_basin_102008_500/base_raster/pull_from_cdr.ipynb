```python
from beak.experimental.utilities.cdr import _download_cdr_files
from beak.experimental.utilities.io import data_folder
```

```python

import requests

import pandas as pd
from pandas import json_normalize
```

```python
token = ""

cma_name = "regional_tusk_great_basin_102008_500" 
url = "https://api.cdr.land/v1/prospectivity/cma?cma_id=ESRI%3A102008_b2f8589dda1a9471873ad66d40bda90a72e7713f09bbee1ac1231ef61c838227__res0_5000_res1_5000_Tungsten"

headers = {
    'accept': 'application/json',
    'Authorization': f'Bearer {token}'
}

response = requests.get(url, headers=headers)

data = response.json()
data = json_normalize(data)

cma = pd.DataFrame(data)
cma
```

```python
cma["download_url"][0]
```

Download template raster

```python
_download_cdr_files(
    download_url=cma["download_url"][0],
    download_file_path=data_folder() / "cdr" / cma_name / "downloads" / "base_raster" / "template_raster.tif",
    verify_ssl=False,
)
```
