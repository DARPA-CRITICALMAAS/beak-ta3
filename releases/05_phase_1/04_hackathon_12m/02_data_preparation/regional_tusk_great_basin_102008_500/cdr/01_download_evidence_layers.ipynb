```python

import requests

import pandas as pd
from pandas import json_normalize

from beak.experimental.cdr import download_evidence_layers
```

# Read data

## From CDR

```python
token = ""
```

### CMA

```python
url = "https://api.cdr.land/v1/prospectivity/cma?cma_id=ESRI%3A102008_b2f8589dda1a9471873ad66d40bda90a72e7713f09bbee1ac1231ef61c838227__res0_5000_res1_5000_Tungsten"

headers = {
    'accept': 'application/json',
    'Authorization': f'Bearer {token}'
}

response = requests.get(url, headers=headers)

data = response.json()
data = json_normalize(data)

df_cma = pd.DataFrame(data)
df_cma
```

### Model run

```python
url = f"https://api.cdr.land/v1/prospectivity/model_run?model_run_id=7da530e4bf5f489d8cea6063a9aad8c1"

headers = {
    'accept': 'application/json',
    'Authorization': f'Bearer {token}'
}

response = requests.get(url, headers=headers)
data = response.json()

df = pd.DataFrame(data)
df
```

### Layers

```python
df_payloads = df.loc["payload"]
df_payloads
```

```python
event = df_payloads["event"]["evidence_layers"]
df_event = json_normalize(event)
df_event
```

```python
cma_id = df_cma.loc[0, "cma_id"]
cma_name = "regional_tusk_great_basin_102008_500"
```

```python
# Some layers do not want to be downloaded via Python commands but work as "click and download"
ignore_layers = ["Geophysics_Gravity_Isostatic", "Geophysics_Gravity_Isostatic_HGM"]

download_evidence_layers(
    data=df_event,
    cma=df_cma.loc[0].to_dict(),
    custom_cma_id=cma_name,
    verify_ssl=False,
    ignore=ignore_layers,
)
```
