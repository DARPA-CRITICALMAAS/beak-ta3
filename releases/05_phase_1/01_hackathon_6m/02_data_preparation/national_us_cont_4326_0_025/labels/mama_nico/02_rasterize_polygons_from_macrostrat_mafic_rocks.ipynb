# Rasterize Macrostrat Mafic Rocks Export
from polygon shapefile to GeoTIFF

Intended to be used as **training labels**.

```python
import numpy as np
import geopandas as gpd

import sys
if sys.version_info < (3, 9):
    from importlib_resources import files
else:
    from importlib.resources import files

from beak.experimental.io import load_raster
```

# Load data

**User definitions**

```python
BASE_PATH = files("beak.data")
PATH_SHAPEFILE = BASE_PATH / "GEOLOGY_LITHOLOGIES" / "RAW" / "Macrostrat_mafic.shp"
PATH_BASE_RASTER = BASE_PATH / "BASE_RASTERS" / "EPSG_4326_RES_0_025_US_CONT.tif"
```

# Create Labels

```python
from beak.experimental.conversions import create_binary_raster

gdf = gpd.read_file(PATH_SHAPEFILE)
base_raster = load_raster(PATH_BASE_RASTER)
out_path = BASE_PATH / "GEOLOGY_LITHOLOGIES" / "PROCESSED" / "EPSG_4326_RES_0_025" / "US_CONT" / "UNIFIED" / "Macrostrat_Mafics.tif"

# Replace out_file=None with out_file=out_path to save the output raster
out_array = create_binary_raster(gdf, base_raster=base_raster, all_touched=False, same_shape=True, fill_negatives=True, out_file=None)

# Check results
np.unique(out_array, return_counts=True)
```
