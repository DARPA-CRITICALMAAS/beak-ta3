# Scaling

**Scaling** all numerical folders within a specified model configuration.<br>
Reads the <code>ROOT_FOLDER</code> and takes the <code>NUMERICAL</code> subfolder for each model configuration.

**User inputs**

```python
BASE_PATH = files("beak.data")

# Points to the "beak" data folder
SPATIAL_DEFINITION = "EPSG_4326_RES_0_025"
SPATIAL_EXTENT = "US_CANADA"
ROOT_FOLDER = BASE_PATH / "LAWLEY22" / "EXPORT" / SPATIAL_DEFINITION / SPATIAL_EXTENT

# Export to current location for testing purposes
CURRENT_DIR = Path(os.getcwd())
PATH_EXPORT = CURRENT_DIR / "EXPORT" / SPATIAL_DEFINITION / SPATIAL_EXTENT

# Select scaling methods
METHODS = ["minmax", "standard"]

# Print folders/subfolders
print("Potential input folders (please select some in the next cell):\n")
for folder in os.listdir(ROOT_FOLDER):
  if os.path.isdir(os.path.join(ROOT_FOLDER, folder)):
    print(folder)
```

```python
SELECTION = ["NUMERICAL", "NUMERICAL_IMPUTED"]
input_folders = [ROOT_FOLDER / folder for folder in SELECTION]
```

```python
methods = ["minmax", "standard"]

for folder in input_folders:
  folder_relative = os.path.relpath(folder, ROOT_FOLDER)
  for method in methods:
    print(f"Processing {folder} with {method} method...")
    output_folder = PATH_EXPORT / str(folder_relative + "_SCALED_" + str.upper(method))
    transformation.scale_raster(folder, output_folder, method)
```
