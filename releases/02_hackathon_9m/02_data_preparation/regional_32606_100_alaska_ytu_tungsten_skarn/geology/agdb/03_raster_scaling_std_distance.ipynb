# Requirements

```python
import os
from importlib_resources import files
from pathlib import Path

from beak.experimental.utilities import transformation
```

# Scaling

**Scaling** all numerical folders within a specified model configuration.<br>
Reads the <code>ROOT_FOLDER</code> and takes the <code>NUMERICAL</code> subfolder within each model configuration.

**User inputs**

```python
BASE_PATH = files("beak.data")
BASE_SPATIAL = "32606_100"
BASE_EXTENT = "alaska_ytu_tungsten_skarn"
BASE_RASTER = BASE_PATH / "BASE_RASTERS" / "EPSG_32606_RES_100_ALASKA_YTU.tif"
BASE_EVIDENCE = "geology"
  
# Export to current location for testing purposes
PATH_ROOT = BASE_PATH / "PROCESSED" / str("regional" + "_" + BASE_SPATIAL + "_" + BASE_EXTENT)
PATH_INPUT = PATH_ROOT / "unified" / BASE_EVIDENCE
PATH_EXPORT = PATH_ROOT / "unified_scaled_std"

# Set export path (select CURRENT_DIR for testing purposes)
CURRENT_DIR = Path(os.getcwd()) / PATH_EXPORT.name

OUT_FOLDER = PATH_EXPORT

print(f"Input folder: {PATH_INPUT}")
print(f"Export folder: {OUT_FOLDER}")

METHOD = "standard"
```

Select subfolders to be scaled.

```python
folders = os.listdir(PATH_INPUT)

for folder in folders:
  if os.path.isdir(os.path.join(PATH_INPUT, folder)):
    print(folder)
  else:
    print(f"No subfolders existing but found {len(folders)} files.")
    break
```

```python
SELECTION = ["distance"]

input_folders = [PATH_INPUT / folder for folder in SELECTION]

print("Selected folders:")
for folder in input_folders:
  print(folder)
```

**Run Scaling** for STD scaler

```python
root_folder = PATH_INPUT.parent

for folder in input_folders:
  folder_relative = os.path.relpath(folder, root_folder)
  print(f"\nProcessing with {METHOD} method...")    
  output_folder = OUT_FOLDER / str(folder_relative)
  print(f"Output folder: {output_folder}")
  transformation.scale_raster(folder, output_folder, METHOD)
```
