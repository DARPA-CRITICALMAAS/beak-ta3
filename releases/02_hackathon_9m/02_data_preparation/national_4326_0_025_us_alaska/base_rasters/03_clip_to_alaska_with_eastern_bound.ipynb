# Case
**Creates binary rasters to be used as templates for data alignment**

Please provide an Esri-shapefile as input.<br>
If there are features you do not want to rasterize, you can create a SQL query to select only the features you want to rasterize.<p>

**Information**

For this specific case, we used the Lawley datacube (2022 release) to create base rasters for 0.025 and 0.015 degree spatial resolution for US and Canada (one base raster for the complete extent).<br>
This way, we ensure that we can mix-up data from different sources. Other base rasters for **national scale** CMAs are based on those and basically clipped or resampled versions of them.


```python
import sys
if sys.version_info < (3, 9):
    from importlib_resources import files
else:
    from importlib.resources import files

from beak.experimental.utilities.raster_processing import clip_raster_file
```

**User definitions**

```python
# Path to base data folder
BASE_PATH = files("beak.data")

# Path to base raster
BASE_RASTER = BASE_PATH / "BASE_RASTERS" / "EPSG_4326_RES_0_025_US_CANADA.tif"

# Path to shapefile to be used for clipping
BASE_SHAPEFILE = BASE_PATH / "RAW" / "boundaries" / "US_States" / "tl_2012_us_state_EPSG_4326.shp"

# Query to filter shapefile, set to None if not necessary
QUERY = "NAME=='Alaska'"

# Restrict eastern bound since Alaska's extent is too large
BOUNDS = (None, None, -129, None)

# Output File
OUT_FILE = BASE_PATH / "BASE_RASTERS" / "EPSG_4326_RES_0_025_US_ALASKA.tif"
```

**Clip Raster**

```python
clipped_raster, clipped_meta = clip_raster_file(
  input_file=BASE_RASTER,
  output_file=OUT_FILE,
  shapefile=BASE_SHAPEFILE,
  query=QUERY,
  bounds=BOUNDS,
 )
```
