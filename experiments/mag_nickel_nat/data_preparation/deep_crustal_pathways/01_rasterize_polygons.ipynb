# Load data

**User definitions**

```python
BASE_PATH = files("beak.data")
PATH_SHAPEFILE = BASE_PATH / "GEOLOGY_DEEP_CRUSTAL_PATHWAYS" / "RAW" / "deep_crustal_pathways.shp"
PATH_BASE_RASTER = BASE_PATH / "BASE_RASTERS" / "EPSG_4326_RES_0_025_US_CONT.tif"

# Export to current location for testing purposes
CURRENT_DIR = Path(os.getcwd())
PATH_EXPORT = CURRENT_DIR / "PROCESSED" 

print(f"Export path: {PATH_EXPORT}")
```

# Create Labels

```python
from beak.utilities.conversions import create_binary_raster

gdf = gpd.read_file(PATH_SHAPEFILE)
base_raster = load_raster(PATH_BASE_RASTER)
out_path = PATH_EXPORT / "EPSG_4326_RES_0_025" / "US_CONT" / "UNIFIED" / "DEEP_CRUSTAL_PATHWAYS_MT_POLS_BEDROSIAN.tif"
labels_array = create_binary_raster(gdf, base_raster=base_raster, all_touched=False, same_shape=True, fill_negatives=True, out_file=out_path)
```
