# Create Training Labels from the Critical Mineral Deposits Database
from shapefile to GeoTIFF


```python
import os
import numpy as np
import geopandas as gpd
from pathlib import Path

import sys
if sys.version_info < (3, 9):
    from importlib_resources import files
else:
    from importlib.resources import files

from beak.utilities.io import load_raster
from beak.utilities.conversions import create_binary_raster
```

# Load data

**User definitions**

```python
# Set base paths and files
BASE_PATH = files("beak.data")
PATH_BASE_RASTER = BASE_PATH / "BASE_RASTERS" / "EPSG_4326_RES_0_025_US_CONT.tif"

# Points file and query to select relevant mineral occurences
PATH_SHAPEFILE = BASE_PATH / "RAW" / "mineral_deposits" / "PorphyryCopper_nat_Dicken16" / "pCu_deps_pros_usgs_80_percent_us_cont.shp"
SQL_QUERY = "country == 'United States'"

# Set the output file
PATH_EXPORT = BASE_PATH / "TRAINING_LABELS" / "PORPHYRY_COPPER_NAT" / str(PATH_BASE_RASTER.stem + "_PCUDEPPRO_80P" + ".tif")

# Overwrite the path for testing purposes
CURRENT_EXPORT = Path(os.getcwd()) / PATH_EXPORT.name
OUT_FILE = PATH_EXPORT
```

# Create Labels
## Complete extent of the dataset
Not saving the results to file

```python
gdf = gpd.read_file(PATH_SHAPEFILE)
base_raster = load_raster(PATH_BASE_RASTER)
out_array = create_binary_raster(gdf, base_raster, query=SQL_QUERY, all_touched=False, same_shape=True, fill_negatives=True, out_file=OUT_FILE)
print(f"Number of positive training labels: {np.sum(out_array==1)}")
```
