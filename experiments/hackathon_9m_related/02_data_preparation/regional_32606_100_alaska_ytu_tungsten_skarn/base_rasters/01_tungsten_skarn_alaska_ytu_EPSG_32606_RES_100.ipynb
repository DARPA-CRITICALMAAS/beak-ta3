# Case
**Creates binary rasters to be used as templates for data alignment**

In this case, we use the high-res geophysical data footprint to create the template raster.

```python
import rasterio
import numpy as np

import sys
if sys.version_info < (3, 9):
    from importlib_resources import files
else:
    from importlib.resources import files

from beak.utilities.raster_processing import clip_raster_file
from beak.utilities.io import read_raster, save_raster
```

**User definitions**

```python
# Path to base data folder
BASE_PATH = files("beak.data")

# Path to footprint raster
INPUT_FILE = BASE_PATH / "RAW" / "geophysics_regional" / "alaska_ytu" / "magnetic" / "Mag_RTP_HGM_YTU.tif"

# Output File
OUT_FILE = BASE_PATH / "BASE_RASTERS" / "EPSG_32606_RES_100_ALASKA_YTU.tif"
```

**Create mask**

```python
# Read raster
raster = rasterio.open(INPUT_FILE)
raster_array = raster.read()

# Define new nodata value
NODATA = -99

# Binarize and set new nodata value
raster_array = np.where(raster_array != raster.nodata, 1, NODATA)

# Save raster
save_raster(path=OUT_FILE, array=raster_array, nodata_value=NODATA, dtype="int8", metadata=raster.meta)
```
