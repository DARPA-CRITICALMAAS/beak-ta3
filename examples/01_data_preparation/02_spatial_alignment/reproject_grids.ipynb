## How to

# Requirements

```python
# Standard libraries
import os

from pathlib import Path
from importlib_resources import files

# Custom modules
from beak.experimental import io, misc, raster_processing
from beak.experimental.io import load_raster
```

# Definitions
**User inputs**

```python
SPATIAL_DEFINITION = "EPSG_4326_RES_0_02"
SPATIAL_EXTENT = "US_CANADA"

# Path to datacube and export folder
BASE_PATH = files("beak.data")
BASE_RASTER = BASE_PATH / "BASE_RASTERS" / "EPSG_102008_RES_2240_US_CANADA.tif"
PATH_INPUT = BASE_PATH / "LAWLEY22" / "EXPORT" / SPATIAL_DEFINITION / SPATIAL_EXTENT

SNAP_RASTER = load_raster(BASE_RASTER)

EPSG_TARGET = 102008
RES_TARGET = int(SNAP_RASTER.transform.a)
CRS_TARGET = SNAP_RASTER.crs

# Export to working location
PATH_EXPORT = BASE_PATH / "LAWLEY22" / "EXPORT" / str("EPSG" + "_" + str(EPSG_TARGET) + "_" + "RES" + "_" + str(RES_TARGET)) / SPATIAL_EXTENT

# Export to example location
CURRENT_DIR = Path(os.getcwd())
EXAMPLE_OUTPUT = CURRENT_DIR / "EXPORT" / str("EPSG" + "_" + str(EPSG_TARGET) + "_" + "RES" + "_" + str(RES_TARGET)) / SPATIAL_EXTENT 

# Set final export option
OUT_FOLDER = PATH_EXPORT
```

**Definitions**

```python
# Input
input_folder = PATH_INPUT

# Output
export_folder = misc.replace_invalid_characters(str(OUT_FOLDER))
export_folder = io.check_path(Path(export_folder))

# Reproject
target_crs = CRS_TARGET
target_resolution = RES_TARGET

print(f"Target CRS: {target_crs}")
print(f"Target resolution: {target_resolution}")
print(f"Input_folder: {input_folder}")
print(f"Export_folder: {export_folder}")
```

# Reprojecting

```python
import multiprocessing as mp

workers = int(mp.cpu_count())
raster_processing.reproject_raster(input_folder, export_folder, target_crs, target_resolution, resampling_mode="auto", include_source=True, n_workers=workers, snap_to_origin=BASE_RASTER)
```
