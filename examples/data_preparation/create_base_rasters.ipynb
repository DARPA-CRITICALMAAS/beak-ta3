# Case
**Creates binary labels in raster format**

The labels are either based on a provided text- or Esri-shapefile.<br>
To spatially align the labels with other data sets, a template raster (base raster) can be provided.


```python
import rasterio
import numpy as np
import pandas as pd
import geopandas as gpd
import os

from pathlib import Path
from typing import List, Tuple, Optional, Union
from rasterio.enums import MergeAlg
    
from rasterio import features

import sys
if sys.version_info < (3, 9):
    from importlib_resources import files
else:
    from importlib.resources import files

from beak.utilities.conversions import create_geodataframe_from_points, create_geodataframe_from_polygons, _rasterize_vector_process, create_binary_raster
from beak.utilities.io import save_raster, check_path, load_raster, load_dataset
```

# Load data

**User definitions**

For a projected **CRS** (such as EPSG:**102008**), use the metric system to provide the resolution (e.g., 2250 m).<br>
For a geographic **CRS** (such as EPSG:**4326**), use degrees to provide the resolution (e.g. 0.025 deg).<br>

**Note**
The input shapefile/geopackage must be in the same CRS as the target base raster.<br>

```python
# Set paths
BASE_PATH = files("beak.data")
PATH_SHAPEFILE = files("beak.data") / "RAW" / "BOUNDARIES" / "US_States" / "tl_2012_us_state_EPSG_102008.shp"
OUT_FILE = "test_base_raster_fn.tif"

# Reference coordinate system code
RESOLUTION = 2250

QUERY = "NAME!='Alaska' & NAME!='Hawaii' & NAME!='Puerto Rico' & NAME!='United States Virgin Islands' & NAME!='American Samoa' & NAME!='Guam' & NAME!='Commonwealth of the Northern Mariana Islands'"
```

Show coverage of shapefile

```python
import matplotlib.pyplot as plt

gdf = gpd.read_file(PATH_SHAPEFILE)
facecolor = "#383e42"
edgecolor = "white"
linewidth = 0.5

if QUERY is not None:
  gdf.query(QUERY).plot(facecolor=facecolor, edgecolor=edgecolor, linewidth=linewidth)
else:
  gdf.plot(facecolor=facecolor, edgecolor=edgecolor, linewidth=linewidth)
  
plt.axis("off")
plt.show()
```

# Create Base Raster

## From shapefile

`fill_negatives`<p>
- True will fill the nodata values with a fill value, commonly 0 (i.e., the extent of the raster is filled completeley with values)<p>
- False will only convert the queried polygons to **one** and the rest of the raster will be **nodata**<p>
**Base rasters** commonly have **nodata** (`fill_negatives=False`) instead of filled values

```python
# Load shapefile and create geodataframe
gdf = gpd.read_file(PATH_SHAPEFILE)

# Save raster
base_array = create_binary_raster(gdf, resolution=RESOLUTION, query=QUERY, all_touched=False, same_shape=True, fill_negatives=False, out_file=OUT_FILE)
```

Show rasterized binary map

```python
from matplotlib.colors import ListedColormap

cmap = ListedColormap(["white", "#383e42"])
cmap.set_bad("black")

fig, ax = plt.subplots(nrows=1, ncols=1)
ax.imshow(base_array, cmap=cmap)
ax.axis("off")
plt.show()
```
