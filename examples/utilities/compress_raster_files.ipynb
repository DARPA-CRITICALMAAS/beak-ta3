Task: Load raster files of a specific format and save them under the same path (overwrite) but with compression.

# Code

Load **libraries**

```python
import rasterio
import gc
from pathlib import Path
from tqdm import tqdm
from beak.utilities.io import create_file_list, save_raster
```

**Set the path** to the folder containing the raster files

```python
FOLDER = Path("S:/Projekte/20230082_DARPA_CriticalMAAS_TA3/Bearbeitung/GitHub/beak-ta3/local/evaluation")
```

**Create** a **list** of **files** to process

```python
file_list = create_file_list(FOLDER, recursive=True)
print(f"Found {len(file_list)} files")
```

**Overwrite** the **files** with the same name but with **compression** applied

```python
for file in tqdm(file_list, desc="Processing files", total=len(file_list), leave=0):
    # Compress
    raster = rasterio.open(file)
    raster_meta = raster.meta.copy()
    raster_array = raster.read(1)
    raster.close()
    
    save_raster(path=file, array=raster_array, metadata=raster_meta)
    
    # Clean-up
    gc.collect()
```
