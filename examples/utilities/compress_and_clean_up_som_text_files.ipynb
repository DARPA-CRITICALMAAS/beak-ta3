# Code

Load **libraries**

```python
import gc
import os
import zipfile

from pathlib import Path
from tqdm import tqdm
from beak.experimental.utilities.io import create_file_list
```

**Set the path** to the folder containing the files and other parameters.

```python
FOLDER = Path(r"S:\Projekte\20230082_DARPA_CriticalMAAS_TA3\Bearbeitung\GitHub\beak-ta3\evaluation")
FILZE_SIZE_THRESHOLD = 0  # in MB
```

**Create** a **list** of **files** to process

```python
file_list = create_file_list(FOLDER, extensions=[".csv", ".txt"], recursive=True)
print(f"Found {len(file_list)} files")
```

**Remove .txt files**

```python
file = file_list[0]
file.name
```

```python
# Get relevant files
txt_file_list = []
for file in file_list:
  if file.name == "result_geo.txt" or file.name == "result_som.txt":
    txt_file_list.append(file)
print(f"Found {len(txt_file_list)} relevant files")

# Clean up
for file in tqdm(txt_file_list):
  os.remove(file)
    
```

**Overwrite** the **files** with the same name but with **compression** applied

```python
# Get relevant files
csv_file_list = []
for file in file_list:
  if file.name == "result_geo.csv" or file.name == "result_som.csv":
    csv_file_list.append(file)
print(f"Found {len(csv_file_list)} relevant files")
```

```python
filtered_file_list = csv_file_list

# Define a function to process each file
def process_file(file):
    file_path = file
    
    # Convert size to MB
    file_size = os.path.getsize(file) / (1024 * 1024)
    
    # Compress
    if file_size > FILZE_SIZE_THRESHOLD:
        zip_file_path = file_path.with_suffix(".zip")
        with zipfile.ZipFile(zip_file_path, "w") as zip_file:
            zip_file.write(file_path, file_path.name, compress_type=zipfile.ZIP_LZMA, compresslevel=5)
        os.remove(file)

    # Clean-up
    gc.collect()

for file in tqdm(filtered_file_list):
    process_file(file)
    
print("Done!")
```
