```python
import numpy as np
import geopandas as gpd

value_columns = ["A", "B"]
height = 5
width = 10
count = len(value_columns)

geodataframe = gpd.GeoDataFrame({"A": np.random.randint(0, 100, size=height), "B": np.random.randint(0, 100, size=height)})
geodataframe
```

```python
args = zip(
    value_columns,
    [geodataframe[column].values for column in value_columns],
    [height] * count,
    [width] * count,

    )

print(*args)
```

```python
args = [
        (
            column,
            geodataframe[column].values,
            height,
            width,
        )
        for column in value_columns
    ]

print(*args)
```
