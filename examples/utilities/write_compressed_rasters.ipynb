```python
import numpy as np, rasterio
size = 16384
chunk = 512

compress_method = 'LZW'              # None or string
compress_num_threads = 'ALL_CPUS'    # None, 'all_cpus', or integer

with rasterio.open('test.tif', 'w', driver='GTiff', nodata=0,
                   width=size, height=size, count=1, dtype=np.float32,
                   tiled=True, blockxsize=256, blockysize=256, 
                   compress=compress_method, num_threads=compress_num_threads) as dst: 
    
    for i in range(0, size, chunk):
        for j in range(0, size, chunk):
            data = np.random.random((chunk, chunk)).astype(np.float32)
            dst.write(data, window=rasterio.windows.Window(i, j, chunk, chunk), indexes=1)
```
