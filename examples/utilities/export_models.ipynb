```python
import pandas as pd
from beak.models.hack_12m_pacree import *
```

```python
all_models = {**national_scale}
all_models.pop("DUMMY_DICTIONARY", None)
all_layers = set()

for model in all_models.values():
    all_layers.update(model.keys())


print(f"Available models: {all_models.keys()}")
```

```python
df = pd.DataFrame(index=list(all_layers), columns=all_models.keys())
for model_name, layers in all_models.items():
    for layer in all_layers:
        df.at[layer, model_name] = "x" if layers.get(layer, False) else "-"
        
df = df.sort_index()
df.to_markdown("export.md", index=True)
```
