```python
from beak.utilities.cdr import _get_cma, _get_evidence_layers, _download_cdr_files, download_evidence_layers
from beak.utilities.io import data_folder
```

```python
json = data_folder() / "cdr" / "json" / "beak_poco.json"
cma_id = "poco_500_test"
```

# Get CMA from json

```python
cma = _get_cma(json)
print(cma.keys())
```

Download template raster

```python
_download_cdr_files(
    download_url=cma["download_url"],
    download_file_path=data_folder() / "cdr" / "cma" / cma_id / "downloads" / "base_rasters" / "template_raster.tif",
    verify_ssl=False,
)
```

Create template raster

```python
import geopandas as gpd
import shapely

from beak.utilities.conversions import create_binary_raster

extent = cma["extent"]
crs = cma["crs"]
resolution = cma["resolution"]

polygon = gpd.GeoDataFrame(
    {"geometry": [shapely.geometry.shape(extent)]},
    crs=crs
)

_, _ = create_binary_raster(
    geodataframe=polygon,
    resolution=500,
    all_touched=False, 
    same_shape=True, 
    fill_negatives=False,
    return_meta=True,
    snap_to_origin=(0,0),   
    out_file="template_raster.tif"
)
```

# Get Evidence Layers

```python
evidence_layers = _get_evidence_layers(json)
```

```python
download_evidence_layers(data=evidence_layers, cma=cma, custom_cma_id=cma_id, verify_ssl=False)
```
